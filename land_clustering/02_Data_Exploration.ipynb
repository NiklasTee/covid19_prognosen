```python
global AuslTeilGeo
Min = 0
Max = 0
def showData(dataCol, csvName, title):
    AuslTeil = pd.concat([karte, dataCol], axis=1)
    AuslTeil.to_csv(csvName, sep=',', encoding='latin-1', index=True)
    AuslTeilGeo = GeoDataFrame(AuslTeil, geometry='geometry')
    #AuslTeilGeo.plot(figsize=(10,10),column=dataCol, legend=True, cmap='RdBu')
    #plt.title(title)
    #plt.axis('off')
    return(AuslTeilGeo)

def MaxMin(x):
    global Max, Min
    if x >= Max[1]:
        return 1
    elif x <= Min[1]:
        return -1
    else: 
        return 0
    
def showMaxMin(data, datacol, title):
    global Max, Min

    Max = (np.max(data[(data[datacol]>0)|(data[datacol]<0)][['GEN', datacol]])).array
    Min = (np.min(data[(data[datacol]>0)|(data[datacol]<0)][['GEN', datacol]])).array

    maxArr = data[datacol].apply(MaxMin)
    #minArr = data[datacol].apply(MaxMin)

    data = pd.concat([data, maxArr], axis=1)
    data = GeoDataFrame(data, geometry='geometry')
    data['maxi'] = maxArr
    display('MAXIMUM',data[data['maxi'] == 1][['GEN',datacol]])
    display('MINIMUM',data[data['maxi'] == -1][['GEN',datacol]])  
    fig, (ax1, ax2) = plt.subplots(ncols=2, sharex=True, sharey=True)

    AuslTeilGeo.plot(ax=ax1,figsize=(10,10),column=datacol, legend=True, )

    data.plot(ax=ax2,column='maxi', legend=True)#figsize=(10,10),column='kmeans', legend=True, cmap='YlGnBu')
    fig.set_figheight(10)
    fig.set_figwidth(20)
    ax1.axis('off')
    ax2.set_title(title)
    ax1.set_title(datacol)

    ax2.axis('off') 
    
    
    
def showBoxplot(datacol, title):
    desc =gdf[datacol]
    print(desc.describe())
    red_square = dict(markerfacecolor='r', marker='s')
    fig5, ax5 = plt.subplots(figsize=(10,3))
    ax5.set_title(title)
    ax5.boxplot(desc, vert=False, flierprops=red_square)
    plt.xlabel(title)
    

    
    
```

```python
#Karte geometry + Nachbarschaft
import pandas as pd
EinWoh = pd.read_csv("01_Einwohner_Flaeche_Data_Pre.csv", encoding='latin-1',  skip_blank_lines=True, skipinitialspace=True, sep=',', index_col=0)
LandK = pd.read_csv("01_landkreise_Data_Pre.csv", encoding='latin-1',  skip_blank_lines=True, skipinitialspace=True, sep=',', index_col=0)
DisM = pd.read_csv("01_distanceMatrix_Data_Pre.csv", encoding='latin-1',  skip_blank_lines=True, skipinitialspace=True, sep=',', index_col=0)
Geo = pd.read_csv("01_Nachbarschaft_Data_Pre.csv", encoding='latin-1',  skip_blank_lines=True, skipinitialspace=True, sep=',', index_col=0)

#display(Geo)
#display(LandK)
#display(DisM)
#display(EinWoh)
colsG = Geo.columns
colsL = LandK.columns
colsD = DisM.columns
colsE = EinWoh.columns
```

```python
A1 = pd.concat([EinWoh, LandK[colsL[2:]].reindex(EinWoh.index)], axis=1)
A2 = pd.concat([A1, DisM[colsD[1:]].reindex(A1.index)], axis=1)
A3 = pd.concat([A2, Geo[colsG[2:]].reindex(A2.index)], axis=1)
A3 = A3.rename(columns={"anteil_bevölkerung_25_bis_45": "anteil_bevölkerung_45_bis_64"})# falscher Titel bearbeiten!
```

Um zugriff auf die Teildaten zu vereinfachen, verwenden wir hirarchische Columns.

```python

def hirarCol(df,rangeL,rangeR,ColName):#Fuktion für hirachical Column. Damit gibt man einen Bereich eines Dataframes einen Column-Namen
    data ={}   
    cols = df.columns[rangeL:rangeR]
    IdDf = df[cols]
    for r in cols:
        data[r] = IdDf[r].array# {cols[0]: IdDf[cols[0]].array, cols[1]: IdDf[cols[1]].array,cols[2]: IdDf[cols[2]].array}
    df = pd.concat({ColName: pd.DataFrame(data)}, axis=1, names=["l1", "l2"])
    return df 
```

```python
################################################################## TO CSV !!
idDf = hirarCol(EinWoh,0,3,'ID')
einWoh_ = hirarCol(EinWoh,3,None,'BEVOE_FLAE')
EinWoh_Hir = pd.concat([idDf, einWoh_.reindex(idDf.index)], axis=1)
EinWoh__ = pd.concat([idDf, einWoh_.reindex(idDf.index)], axis=1)
```

```python
################################################################## TO CSV !!
landK_ = hirarCol(LandK,2,None,'LandKreise')
landK_Hir = pd.concat([idDf, landK_.reindex(idDf.index)], axis=1)
#landK_Hir
```

```python
################################################################## TO CSV !!
DistMa_ = hirarCol(DisM,1,None,'Dist_Matrix')
DistMa_Hir = pd.concat([idDf, DistMa_.reindex(idDf.index)], axis=1)
#DistMa_Hir
```

```python
################################################################## TO CSV !!
Geo_ = hirarCol(Geo,2,None,'Geo_Nachbar')
Geo_Hir = pd.concat([idDf, Geo_], axis=1)
#Geo_Hir
```

```python
DataConsold_Hir = pd.concat([EinWoh_Hir, landK_,DistMa_,Geo_], axis=1)
#DataConsold_Hir
```

```python
EinWoh_Hir.to_csv(r'02_EinWoh_Hir.csv', sep=',', encoding='latin-1', index=True)
landK_Hir.to_csv(r'02_landK_Hir.csv', sep=',', encoding='latin-1', index=True)
DistMa_Hir.to_csv(r'02_DistMa_Hir.csv', sep=',', encoding='latin-1', index=True)
Geo_Hir.to_csv(r'02_Geo_Hir.csv', sep=',', encoding='latin-1', index=True)
DataConsold_Hir.to_csv('02_DataConsold_Hir.csv', sep=',', encoding='latin-1', index=True)
A3.to_csv('02_DataConsold_.csv', sep=',', encoding='latin-1', index=True)
A3 = pd.read_csv("02_DataConsold_.csv", encoding='latin-1',  skip_blank_lines=True, skipinitialspace=True, sep=',', index_col=0)
A3.head()
```
